```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
df = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
df
```

```python
transactions = []
for i in range(0, 7499):
    transactions.append([str(df.values[i,j]) for j in range(0, 19)])
```

```python
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
results = list(rules)
results
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['ITEM_1', 'ITEM_2', 'Support', 'Confidence', 'Lift'])
```

```python
clean = resultsinDataFrame.nlargest(n = 15, columns = 'Support')
clean.sort_values(by=['Support','Confidence','Lift'],ascending=False)
```
